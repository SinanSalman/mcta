```python
import mcta
import mcta_rw
import timeit
base = 'AD2'
Settings, GeoJSON_Map = mcta_rw.LoadDataFiles(base)
lengths, lanes, FFS, P = mcta.SetupMCTA(GeoJSON_Map, Verbose=False)

def A():
    mcta.SolveMCTA(lengths, lanes, P, \
                   VehiclesCount=GeoJSON_Map['mcta_json']['VehiclesCountEst'],\
                   Objectives=['D','K','C','PI'], SkipChecksForSpeed=False)
```

```python
%%timeit
A()
```

```python
%prun b=timeit.timeit(A,number=25)
```
